# Get City Weather Data

```python
# Import the dependencies.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Use the citipy module to determine city based on latitude and longitude.
from citipy import citipy

# importing requests for API call
import requests

# importing api key saved in config file
from config import weather_api_key

```

```python
# random latitudes and longitudes comboms
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)

combo_lats_lngs = zip(lats, lngs)

combo_lats_lngs
```

```python
# Add the latitudes and longitudes to a list (1500 items)
coordinates = list(combo_lats_lngs)

coordinates
```

```python
# creating a true list of cities without duplicates

# a dictionary to store the names about to appended
cities = []

# Identifying nearest city for each lat and long combo

for coordinate in coordinates:
    city = ctp.nearest_city(coordinate[0], coordinate[1]).city_name
    
    # if city is unique, then it is added, no so duplicates
    if city not in cities:
        cities.append(city)
        
# print the lenght of the cities
len(cities)
```

```python
list(cities)
```

```python
url = f"http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID={weather_api_key}"
```

```python
# declaring empty list to collect data from api

city_data = []

# printing begining of logging

print("Beginning Data Retreival           ")
print("------------------------------------")

# creating counters

record_count = 1

set_count = 1

# setting records here so when we loop arounf 50 sets of cities at a time, we can find and keep track
# of the number of loops, as the records themselves will be looped through

for i, city in enumerate(cities):
    print(i, city)
    
#     grouping cities by sets of 50 to log and check
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
#   creating endpoint url for each city
#   since the city names have space in between them, this replaces the space with a plus for the url
    city_url = f"{url}&q={city.replace(' ','+')}"
    
#   log url record and set numbers to check processing
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    
#   add 1 to every record added 
    record_count += 1
    
    try:
        # Parse the JSON and retrieve data, assign a variable first
        city_weather = requests.get(city_url).json()
        # Parse out the needed data, separate
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        # Convert the date to ISO standard
        city_date = datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
        # Append the city information into city_data list, precursor to a dataframe
        city_data.append({"City": city.title(),
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Country": city_country,
                          "Date": city_date})

# If an error is experienced, skip the city.
    except:
        print("City not found. Skipping...")
#       countinues the loop without stopping
        pass

# Indicate that Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
    
    
    
    
    
len(city_data)
```

```python
# converting to a dataframe
city_data_df = pd.DataFrame(city_data)
city_data_df.head(10)

# reordering columns
new_cols = ["City", "Country", "Date", "Lat", "Lng", "Max Temp", "Humidity", "Cloudiness", "Wind Speed"]
city_data_df = city_data_df[new_cols]

city_data_df

# saving the dataframe as a csv file in the folder

convert_df2csv = "weather_data/cities.csv"

city_data_df.to_csv(convert_df2csv, index_label = "City_ID")

city_data_df = pd.read_csv("weather_data/cities.csv")

city_data_df.head(10)
```

```python
# adding scatter plots of latitude vs various weather indicies

lats_data = city_data_df['Lat']
maxtemp_data = city_data_df['Max Temp']
humidity_data = city_data_df["Humidity"]
cloudiness_data = city_data_df["Cloudiness"]
wind_speed_data = city_data_df["Wind Speed"]

wind_speed_data.head(10)
```

```python
# importing time module
import time 

# scatter plot build lats vs max temp

plt.scatter(lats_data, 
            maxtemp_data,
            edgecolor='black',
            linewidths=1,
            marker="o",
            label="Cities")

plt.title(f"City Latitude vs Max Temperature {time.strftime('%x')}")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")

# saving plot
plt.savefig("weather_data/Fig1.png")

# show plot
plt.show()
```

```python
# humiditiy

plt.scatter(lats_data, 
            humidity_data,
            edgecolor='black',
            linewidths=1,
            marker="o",
            label="Cities",
            alpha=0.8)

plt.title(f"City Latitude vs Humidity {time.strftime('%x')}")
plt.xlabel("Latitude")
plt.ylabel("Humidity (%)")

plt.grid(True)

# Save the figure.
plt.savefig("weather_data/Fig2.png")

# Show plot.
plt.show()
```
