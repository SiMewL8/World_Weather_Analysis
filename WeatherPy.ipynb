```python
# Import the dependencies.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# random latitudes and longitudes comboms
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)

combo_lats_lngs = zip(lats, lngs)

combo_lats_lngs
```

```python
# Create a practice set of random latitude and longitude combinations.
x = [25.12903645, 25.92017388, 26.62509167, -59.98969384, 37.30571269]
y = [-67.59741259, 11.09532135, 74.84233102, -76.89176677, -61.13376282]
test_coord = zip(x, y)
```

```python
# Use the tuple() function to display the latitude and longitude combinations.
for coordinate in test_coord:
    print(coordinate[0], coordinate[1])

# loops in x and y values in a tuple()
# [(25.12903645,-67.59741259),(25.92017388,11.09532135), (), (), ]
```

```python
# Add the latitudes and longitudes to a list (1500 items)
coordinates = list(combo_lats_lngs)

coordinates
```

```python
# Use the citipy module to determine city based on latitude and longitude.
from citipy import citipy as ctp
```

```python
# Use the tuple() function to display the latitude and longitude combinations.
for coordinate in coordinates:
    print(ctp.nearest_city(coordinate[0], coordinate[1]).city_name,
          ctp.nearest_city(coordinate[0], coordinate[1]).country_code)

# iterate thro test_coord
# use citipy.nearest_city()
# for loop will iterate the values in test_coord thro the index "coordinate[]"
# then will provide city name
# then will provide country initials 
```

```python
# creating a true list of cities without duplicates

# a dictionary to store the names about to appended
cities = []

# Identifying nearest city for each lat and long combo

for coordinate in coordinates:
    city = ctp.nearest_city(coordinate[0], coordinate[1]).city_name
    
    # if city is unique, then it is added, no so duplicates
    if city not in cities:
        cities.append(city)
        
# print the lenght of the cities
len(cities)
```
